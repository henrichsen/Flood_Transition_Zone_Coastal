# Run CFE and Prepare File for GeoClaw 
This file will run CFE for a given catchment on either Hurricane Harvey or Ike, and trim the output file to the time range of the GeoClaw simulation. It requires forcing file and config files as input, and requires the user to set the catchment name and storm name.. The output_name (name of the cfe output file), cfg_file (path to configuration file to run cfe), and the GeoClaw_input_file_name (name of the file to be used as input for the GeoClaw model) variables can be changed if desired.

```python
#Set up environment
import time
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, './cfe')
import bmi_cfe
```

In this cell, enter the catchment name (*catch* variable) and the storm name (*storm* variable). With the included example data, *catch* can be "Double", "Cedar", or "Oyster" and *storm* can be "Ike" or "Harvey".

```python
# Set which catchment and storm
catch = "Cedar" #This can be Double, Cedar, or Oyster
storm = "Harvey" #This can be Harvey or Ike

# Set the path to the config file
cfg_file = './config/cat-'+catch+'_config_cfe_'+storm+'.json'

# Set file names for CFE output and adjusted GeoClaw input files
output_name = 'CFE_output_'+storm+'_'+catch+'.csv'
GeoClaw_input_file_name = 'output/forGeoClaw/adjusted_'+output_name
```

```python
#Create instance of cfe
cfe_instance = bmi_cfe.BMI_CFE()

#Initialize for catchment subregion 
cfe_instance.initialize(cfg_file=cfg_file)

#Open forcing file in location described in config file 
with open(cfe_instance.forcing_file, 'r') as f:
    df_forcing = pd.read_csv(f)

# Set up dataframe for outputs from CFE
cfe_outputs=cfe_instance.get_output_var_names()
#print(cfe_outputs)
cfe_output_lists = {output:[] for output in cfe_outputs}
out_df = pd.DataFrame(columns = cfe_outputs)
```

```python
# Run CFE for each time step in the forcing file
for i in range(0, df_forcing['precip_rate'].shape[0]):

    hour_of_simulation = i
        
    precip = df_forcing.loc[hour_of_simulation, 'precip_rate'] 
    TMP_2maboveground = df_forcing.loc[hour_of_simulation, 'TMP_2maboveground']
        
    # Set potential evapotranspiration to zero
    # Note that adding PET did not notably influence output values of CFE, so it is removed for simplicity
    cfe_instance.set_value('water_potential_evaporation_flux', 0)
    # Set precip flux value
    cfe_instance.set_value('atmosphere_water__time_integral_of_precipitation_mass_flux', precip)
    cfe_instance.update()
             
    for output in cfe_outputs:
    
        cfe_output_lists[output].append(cfe_instance.get_value(output))
        
        out = cfe_instance.get_value(output)
        out_df.loc[i, output]= out
        out_df.loc[i,"date_time"] = df_forcing.loc[i, "time"]
        
# Print first five rows of CFE result
print(out_df.head())

# Write results to file
out_df.to_csv('output/'+output_name,sep=',', header=True)
```

You should now have a CFE output file in the output folder. 

The next step is to close the CFE instance. You can check the residuals for each of the mass balances to ensure they are close to zero. 

```python
#close cfe and print mass balance
cfe_instance.finalize(print_mass_balance=True)
```

```python
#plot results
istart_plot=100
iend_plot=8970
x = list(range(istart_plot, iend_plot))
for output in cfe_outputs:
    plt.plot(x, cfe_output_lists[output][istart_plot:iend_plot], label=output)
    plt.title(output)
    plt.legend()
    plt.show()
    plt.close()
```

## Trim CFE Output to GeoClaw timeframe
The code block below trims the CFE output dataset to only include the time period of the GeoClaw model, and writes a new trimmed output file to the path defined earlier with the *GeoClaw_input_file_name* variable. The dates included here are specific to the GeoClaw models for Ike and Harvey, but the code could be modified for other storm dates if desired. 

```python
#Read the cfe output file into a new dataframe
df = pd.read_csv('output/'+output_name)
#print(df.head())
#print(out_df.head())

#Find start and end times of the GeoClaw simulation in the cfe output dataframe
if storm == 'Ike':
    start = df[df.date_time == "9/10/2008 7:00"].index[0]
    end = df[df.date_time == "9/14/2008 7:00"].index[0]
elif storm == 'Harvey':
    start = df[df.date_time == "8/23/2017 6:00"].index[0]
    end = df[df.date_time == "8/31/2017 6:00"].index[0]
else:
    print('Error in setting up output file for GeoClaw timeline: Need to set storm as Harvey or Ike')


    
#Trim dataframe to GeoClaw simulation time
df = df.iloc[: , 1:]
df2 = df.iloc[start:end]

#Write trimmed dataframe to csv to read into GeoClaw
df2.to_csv(GeoClaw_input_file_name, index=False)
```

You should now have a file populated in the output/forGeoClaw folder.
